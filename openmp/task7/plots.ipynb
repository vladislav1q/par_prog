```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(a,x,y):
    return a[int(y)][int(x)]
```

```python
N = 128
M = 128
f = open('results.txt', 'r')
for line in f.readlines():
    results = line.split(' ')
    
f.close()

temperatures = []
for i in results[:-1]:
    temperatures.append(float(i))
        
temperatures = np.reshape(temperatures,(N,M))

x = np.linspace(1, M, M)
y = np.linspace(N, 1, N)

X, Y = np.meshgrid(x, y)

plt.figure(figsize=(12,12))


plt.scatter(X,Y,c=(temperatures))
# plt.contour(X, Y, temperatures, 20, cmap='RdGy')
```
