```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
f = open('results.txt', 'r')
for line in f.readlines():
    results = line.split(' ')
f.close()
results = results[:-1]

sequent = []
static = []
dynamic = []
guided = []
for i in np.arange(0,10):
    sequent.append(float(results[4*i]))
    static.append(float(results[4*i+1]))
    dynamic.append(float(results[4*i+2]))
    guided.append(float(results[4*i+3]))

plt.figure(figsize=(12,8))
plt.plot(np.arange(0,10), sequent, label='sequent')
plt.plot(np.arange(0,10), static, label='static')
plt.plot(np.arange(0,10), dynamic, label='dynamic')
plt.plot(np.arange(0,10), guided, label='guided')
plt.xlabel('Number of threads')
plt.ylabel('Time of execution for 2 mults 1000*1000')
```
